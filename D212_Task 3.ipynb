```python
#import packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#import data file & preview data 
df = pd.read_csv('/Users/Amanda.Hartzler/Desktop/Data_Analytics_Masters/D212/Task 3/teleco_market_basket.csv')
df.head()
```

```python
#Remove empty rows
df2 = df.dropna(how='all')
df2.head()
```

```python
df2.info()
```

```python
df2.shape
```

```python
#Convert dataframe into a list of lists by running two loops and columns as items (Kamara, 2022)
transactions = []
for i in range (0, len(df2)):
    transactions.append([str(df2.values[i,j]) for j in range(0,len(df2.columns))])
    
#Transform the list of lists into a Numpy array for analysis (Kamara, 2022)
TE = TransactionEncoder()
array = TE.fit(transactions).transform(transactions)
```

```python
#Create dataframe where rows are logical and columns are purchased items (Kamara, 2022)
clean_df = pd.DataFrame(array, columns=TE.columns_)
clean_df
```

```python
#View column names
for col in clean_df.columns:
    print(col)
```

```python
#Drop empty column (nan)
df_clean = clean_df.drop(['nan'], axis=1)
```

```python
#View new column names
for col in df_clean.columns:
    print(col)
```

```python
#Copy of cleaned dataset
df_clean.to_csv('/Users/Amanda.Hartzler/Desktop/D212_PA_Task3.csv')
```

```python
#Apply apriori algorithm (Kamara, 2022)
a_rules = apriori(df_clean, min_support = 0.015, use_colnames = True)
a_rules.head()
```

```python
a_rules_results = list(a_rules) #(Kamara, 2022)
a_rules_results
```

```python
#Association rules (Kamara, 2022)
ass_r = association_rules(a_rules, metric = 'lift', min_threshold = 1)
ass_r.head(10)
```

```python
#Top three rules (Kamara, 2022)
top_3_rules = ass_r.sort_values('lift', ascending = False).head(3)
top_3_rules
```
